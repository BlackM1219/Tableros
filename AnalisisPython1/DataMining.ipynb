```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA 
```

```python
data = sns.load_dataset('iris')
data
```

```python
pca_model = PCA(n_components=2)  # Definir el modelo PCA para reducir a 2 componentes
data_pca = pca_model.fit_transform(data.iloc[:, :-1])  # Aplicar PCA a las características, excluyendo la columna de etiquetas
data_pca
```

```python
plt.scatter(data_pca[:, 0], data_pca[:, 1], c=data['species'].astype('category').cat.codes, cmap='viridis', alpha=0.5)
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('PCA de Iris Dataset')
plt.colorbar(label='Especies')
plt.show()
```

```python
#trasnformacionde variables
sns.displot(data=data, x='sepal_length', hue='species', kde='True')  # Gráfico de densidad para la longitud del sépalo por especie
plt.show()  # Mostrar el gráfico
```

```python
#transformacion logaritmica
data['sepal_length_log'] = np.log(data['sepal_length'])  # Aplicar transformación logarítmica a la longitud del sépalo
sns.displot(data=data, x='sepal_length_log', hue='species', kde='True')  # Gráfico de densidad para la longitud del sépalo transformada logarítmicamente por especie
plt.show()  # Mostrar el gráfico
```

```python
from scipy.stats import pearsonr
```

```python
#calcular la correlacion de pearson entre las variables
correlacion, _ = pearsonr(data['sepal_length'], data['sepal_width'])
```

```python
print(f'Correlación de Pearson entre la longitud y el ancho del sépalo: {correlacion:.2f}')  # Imprimir la correlación de Pearson
```

```python
# analisis de cluster, separa los datos en grupos de acuerdo a sus similitudes
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3, random_state=42)  # Definir el modelo KMeans con 3 clusters
data['cluster'] = model.fit_predict(data.iloc[:, :-2])  # Ajustar el modelo y predecir los clusters, excluyendo la columna de etiquetas
data
```

```python
sns.scatterplot(data= data, x='sepal_length', y='petal_width', hue='cluster', palette='viridis')  # Gráfico de dispersión de los clusters
plt.title('Clusters de Iris Dataset')
plt.xlabel('Longitud del Sépalo')
plt.ylabel('Ancho del Pétalo')
plt.show()  # Mostrar el gráfico
```

```python
# metodo de zscore para detectar outliers
from scipy import stats
```

```python
#calcular valores z 
data['z_score_sepal_length'] = np.abs((data['sepal_length'] - data['sepal_length'].mean()) / data['sepal_length'].std())
data
```

```python
outliers = data[data['z_score_sepal_length'] > 2]  # Filtrar los outliers basados en el valor Z
outliers
```

```python
sns.scatterplot(data= data, x='sepal_length', y='petal_length', hue='species', palette='viridis') # Gráfico de dispersión de la longitud del sépalo y la longitud del pétalo por especie
sns.scatterplot(data=outliers, x='sepal_length', y='petal_length', color='red', label='Outliers')  # Añadir los outliers al gráfico
plt.title('Outliers en el Iris Dataset')
```
